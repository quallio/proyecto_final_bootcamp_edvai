```python
import pandas as pd
import pickle

data = pd.read_csv("data/customer_dataset.csv", sep=',') 
```

```python
data.shape
```

```python
# Antes que nada elimino aquellas rows que identifiqué en el notebook anterior, que son outliers.
# En realidad es una sola row.

data = data.loc[data["transactionAmount"] < 80]
```

```python
data.shape
```

```python
probando = data.copy()
```

```python
# Para la columna orderAmount

data['orderAmount'], saved_bins_order = pd.qcut(data['orderAmount'], q=5, duplicates='drop', retbins=True)
```

```python
with open('data/saved_bins_order.pickle', 'wb') as handle:
  pickle.dump(saved_bins_order, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Para la columna transactionAmount

data['transactionAmount'], saved_bins_transaction = pd.qcut(data['transactionAmount'], q=3, duplicates='drop', retbins=True)
```

```python
with open('data/saved_bins_transaction.pickle', 'wb') as handle:
  pickle.dump(saved_bins_transaction, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Preparación de los datos

from funpymodeling import status

status(data)
```

```python
data['paymentMethodIssuer'].value_counts()
```

```python
# Reemplazar un valor a la vez
weird_payment_method_issuer = ["B", "e", "c", "r", " ", "n", "x", "o", "a", "p"]
# Estos son todos los RAROS que he encontrado en el paso anterior

for payment_method_issuer in weird_payment_method_issuer:
    data['paymentMethodIssuer'] = data['paymentMethodIssuer'].replace(payment_method_issuer, 'weird')
```

```python
data['paymentMethodIssuer'].value_counts()
```

```python
data['paymentMethodProvider'].value_counts()
```

```python
# Agrupo los JCB de 16 y 15 digit, y lo mismo con los VISA. Así quedan 1 VISA y 1 JCB

VISA = ["VISA 16 digit", "VISA 13 digit"]

for payment_method_provider in VISA:
    data['paymentMethodProvider'] = data['paymentMethodProvider'].replace(payment_method_provider, 'VISA')
```

```python
JCB = ["JCB 16 digit", "JCB 15 digit"]

for payment_method_provider in JCB:
    data['paymentMethodProvider'] = data['paymentMethodProvider'].replace(payment_method_provider, 'JCB')
```

```python
data['paymentMethodProvider'].value_counts()
```

```python
data['paymentMethodType'].value_counts()
```

```python
# Agrupo los tipos de pago en 2: card o other

other_payment_method = ["apple pay", "paypal", "bitcoin"]

for payment_method_type in other_payment_method:
    data['paymentMethodType'] = data['paymentMethodType'].replace(payment_method_type, 'other')
```

```python
data['paymentMethodType'].value_counts()
```

```python
status(data)
```

```python
# - fraudulent
# Vamos a hacer una especie de semáforo:

# False = Green le asignaremos el valor numérico 0
# True = Red le asignaremos el valor numérico 1
# Nan = Yellow le asignaremos el valor numérico 2
```

```python
data['fraudulent']=data['fraudulent'].fillna(value="warning")
```

```python
data["fraudulent"].value_counts()
```

```python
# Nos aseguramos que no exista ningún tipo de conflicto por falta de compatibilidad entre str y bool de True/False.

data['fraudulent']=data['fraudulent'].astype(str)
```

```python
class_map = {'False': 0, 'True': 1, 'warning': 2}
data['fraudulent'] = data['fraudulent'].map(class_map)
```

```python
data["fraudulent"].value_counts()
```

```python
data.head(3)
```

```python
# Custom tratamiento de datos faltantes

# Lo hacemos para columnas categóricas, que en nuestro caso la única categórica con datos faltantes es orderAmount.
```

```python
status(data)
```

```python
data['orderAmount'].value_counts()
```

```python
data['orderAmount']=data['orderAmount'].cat.add_categories("desconocido")
data['orderAmount']=data['orderAmount'].fillna(value="desconocido")
```

```python
data['orderAmount'].value_counts()
```

```python
# Guardamos el DataFrame en un csv antes de hacerle el OHE y eliminar la columna de fraudulent

filename0 = "data/customer_dataset_preparado_sin_ohe.csv"
data.to_csv(filename0, index = False)
```

```python
# ONE HOT ENCODING

data_ohe = data.copy()
```

```python
# Eliminamos la columna de fraudulent

data_ohe_without_fraudulent = data_ohe.drop(["fraudulent"], axis=1)
```

```python
# Hacemos el get_dummies y aquellos valores NULOS tendrán una columna para cada variable

data_ohe_without_fraudulent=pd.get_dummies(data_ohe_without_fraudulent, dummy_na=True)
```

```python
# Guardar el nombre de las columnas con el nombre: categories_ohe_without_fraudulent.pickle

with open('data/categories_ohe_without_fraudulent.pickle', 'wb') as handle:
    pickle.dump(data_ohe_without_fraudulent.columns, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Guardamos dataset

filename = "data/ohe_customer_dataset.csv"
data_ohe_without_fraudulent.to_csv(filename, index = False)
```
